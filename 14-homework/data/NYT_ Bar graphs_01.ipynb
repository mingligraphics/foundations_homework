```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_csv("social-spending.csv")
df
```

```python
df['Social Spending'] = df['Social Spending'].str.extract(r"(.*)%")
df['Social Spending'] = df['Social Spending'].astype(float)
df
```

```python
def build_colors(row):
    if row['Country'] == 'O.E.C.D. avg.':
        return 'blue'
    elif row['Country'] == 'Finland':
        return 'blue'
    elif row['Country'] == 'Sweden':
        return 'blue'
    elif row['Country'] == 'Finland':
        return 'blue'
    elif row['Country'] == 'United States':
        return 'blue'
    else:
        return 'lightblue'

colors = df.reset_index().apply(build_colors, axis=1)
```

```python
df.sort_values(by='Social Spending').plot(x='Country',kind='barh',color=colors, figsize=(9,8), width=0.8, title='Public Social Spending as a share of G.D.P')
plt.savefig("NYT_Bar01.pdf")
```
