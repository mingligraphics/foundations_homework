```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_csv("diversity-by-quarter.csv")
df
```

```python
ax = df[df['toa_group']=='all-permanent'].plot(x='quarter', y='pct_minority',style='o-', fillstyle='none')
ax.set_ylim(0,50)
df[df['toa_group']=='midlevel-appointees'].plot(ax=ax, x='quarter', y='pct_minority',style='o-', fillstyle='none')
ax.grid(linestyle='--', color='lightgrey')
ax.tick_params(which='both', # Options for both major and minor ticks
                top='off', # turn off top ticks
                left='off', # turn off left ticks
                right='off',  # turn off right ticks
                bottom='off')
plt.savefig("Buzzfeed_01.pdf")
```
