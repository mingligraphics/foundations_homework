```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df = pd.read_csv("order-of-australia.csv")
```

```python
df['percent_female'] = 100 - df['percent_male']
```

```python
df
```

```python
fig, ax = plt.subplots()
colors=['blue','lightblue']
df.sort_values('field', ascending=False).plot(x='field', kind='barh', figsize=(10, 20), color=colors, ax=ax)
ax.grid(which='major', linestyle='-', linewidth='0.5', color='lightgrey')
plt.savefig("guardian01.pdf")
```
