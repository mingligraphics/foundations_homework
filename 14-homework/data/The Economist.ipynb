```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df = pd.read_csv("bank-profits.csv")
```

```python
df
```

```python
colors=['darkgreen','green','darkblue','blue','lightgreen','lightblue']
df[['year-2007','year-2011']].plot(kind='pie', subplots=True, labels=df.region, colors=colors)
plt.savefig("economist01.pdf")
```

```python
fig, ax = plt.subplots()
ax.hlines(xmin=df['year-2007'], xmax=df['year-2011'], y=df.index)
ax.scatter(x=df['year-2011'], y=df.index)
ax.set_yticks(df.index)
ax.yaxis.set_ticklabels(df['region'])

plt.savefig("economist02.pdf")
```
