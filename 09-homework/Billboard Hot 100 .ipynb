```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.billboard.com/charts/hot-100'
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
songs = doc.find_all(class_='chart-list-item')
```

```python
rows = []
for song in songs:
    row = {}
    row['rank'] = song.find(class_='chart-list-item__rank').text.replace("\n","")
    row['title'] = song.find(class_ = 'chart-list-item__title').text.replace("\n","")
    row['name'] = song.find(class_='chart-list-item__artist').text.replace("\n","")
    rows.append(row)
    
```
