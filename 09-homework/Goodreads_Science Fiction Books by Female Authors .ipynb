```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = "https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors"
```

```python
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
books = doc.find_all('tr')
```

```python
rows = []
for book in books:
    row = {}
    row['Rank'] = book.find(class_="number").text
    row['Title'] = book.find(class_="bookTitle").text.replace("\n","")
    row['Author'] = book.find(class_="authorName").text
    row['Score'] = book.find(class_="smallText uitext").find('a').text
    row['Votes'] = book.find(class_="smallText uitext").find_all('a')[1].text
    row['Rating'] = book.find(class_="minirating").text
    rows.append(row)
```

```python
df = pd.DataFrame(rows)
```

```python
df = df[['Rank','Title','Author','Score','Votes','Rating']]
```
