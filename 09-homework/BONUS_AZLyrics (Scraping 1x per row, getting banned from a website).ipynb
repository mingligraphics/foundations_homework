```python
import pandas as pd
import requests

from bs4 import BeautifulSoup
```

```python
url = "https://www.azlyrics.com/r/rem.html"
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
rows = []
album = doc.find(id="listAlbum")
for song in album.find_all('a'):
    row = {}
    row['Title'] = song.text
    row['URL'] = song['href']
    rows.append(row)
    
```

```python
rows
```

```python
df = pd.DataFrame(rows)
```

```python
df['URL'] = df['URL'].astype(str)
df['URL'] = df.URL.str.extract(r"(/lyrics.*)")
```

```python
df.head()
```

```python
df.to_csv("AZLyrics.csv", index=False)
```

## Part II

```python
def scrape_page(row):
    url = f"https://www.azlyrics.com{row['URL']}"
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    page = {}
    page['verses'] = doc.find(class_= "col-xs-12 col-lg-8 text-center").find_all('div')[6].text.strip()
    return pd.Series(page)
```

```python
scrape_page(df.loc[5])
```

```python
scraped_df = df.apply(scrape_page, axis=1)
```

```python
scraped_df = df.merge(scraped_df, left_index=True, right_index=True)
scraped_df.head()
```
