```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
rows = []
for page_num in range(1,11):
    url = f"https://www.epicurious.com/search/cucumbers?page={page_num}"
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    dishes = doc.find_all("article")    
    for dish in dishes:
        row = {}
        try:            
            row['Tag'] = dish.find(class_="tag").text
        except:
            pass
        try:
            row['Title'] = dish.find(class_="hed").text
        except:
            pass
        try:
            row['Summary'] = dish.find(class_="dek").text
        except:
            pass
        try:
            row['Rating'] = dish.find(itemprop="ratingValue").text
        except:
            pass
        try:
            row['Would Make Again'] = dish.find(class_="make-again-percentage").text
        except:
            pass
        try:
            row['Link'] = dish.find(class_="hed").find('a')['href']
        except:
            pass
        rows.append(row)
    print(rows)
```

```python
df = pd.DataFrame(rows)
```

```python
df = df[['Tag','Title','Summary','Rating','Would Make Again','Link']]
df = df.dropna(subset=["Title"])
df.head()
```

```python
df.to_csv("Epicurious.csv", index=False)
```

# Part II

```python
df_recipes = df[df.Tag == 'recipe']
```

```python
df_recipes.head()
```

```python
def scrape_page(row):
    url = "https://www.epicurious.com" + row['Link']
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)

    page = {}
    
    Ingredients = doc.find_all(class_="ingredient")
    page['Ingredients'] = '\n'.join([ingredient.text for ingredient in Ingredients])

    Directions = doc.find_all(class_="preparation-step")
    page['Directions'] = '\n'.join([direction.text for direction in Directions])

    Tags = doc.find_all(itemprop="recipeCategory")
    page['Tags'] = '\n'.join([tag.text for tag in Tags])

    return pd.Series(page)
```

```python
scrape_page(df_recipes.loc[1])
```

```python
scraped_df = df_recipes.apply(scrape_page,axis=1)
```

```python
df_recipes.merge(scraped_df, left_index=True, right_index=True)
```

```python
df_recipes.to_csv("Epicurious_recipes.csv", index=False)
```
